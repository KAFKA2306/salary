```python
import pandas as pd
import glob

# CSVファイルのパスを指定
path = 'C:/Users/100ca/Documents/PyCode/scraping/data/*.csv'

# パスにマッチするすべてのファイルを取得
csv_files = glob.glob(path)

# 各CSVファイルに対して処理
for file in csv_files:
    # CSVファイルを読み込み
    df = pd.read_csv(file)
    
    # 'salary_25_29'列で降順にソート
    df = df.sort_values(['salary_25_29','salary'], ascending=False)
    display(df)
    # 同じファイル名でCSVファイルを上書き保存
    df.to_csv(file, index=False)
```
